```python
from torch import nn
```

```python
# 创建模型
linear = nn.Linear(3,5)
# 查看参数
print(linear.weight)
print(linear.bias)
```

```python
# 更新参数
nn.init.normal_(linear.weight)    # 正态分布初始化
```

```python
nn.init.uniform_(linear.weight)     # 均匀分布初始化
```

```python
nn.init.xavier_uniform_(linear.weight)  # Xavier均匀分布初始化
```

```python
nn.init.xavier_normal_(linear.weight)    # Xavier正态分布初始化
```

```python
nn.init.kaiming_uniform_(linear.weight)  # Kaiming均匀分布初始化
```

```python
nn.init.kaiming_normal_(linear.weight)    # Kaiming正态分布初始化
```

```python
nn.init.ones_(linear.bias)    # 全1初始化
nn.init.zeros_(linear.bias)    # 全0初始化
nn.init.constant_(linear.bias, 66)
```
